```python
df = df.rename(columns={'sl_no': 'N', "ssc_p": "seconday_educ_per", "ssc_b": "board_of_educ", "hsc_p": "high_second_educ_per", "hsc_b": "board_of_high_educ", "hsc_s": "special_high_sec_educ", "degree_t": "degree_field"})
```

```python
gender_counts = df["gender"].value_counts()

gender_counts.plot(kind="bar")
plt.title("Gender Counts")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
percentages = df["gender"].value_counts(normalize=True) * 100
percentages.plot(kind='bar')
plt.title('Gender Percentages')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
percentages = df["specialisation"].value_counts(normalize=True) * 100
percentages.plot(kind='bar')
plt.title('Specialization Percentages')
plt.xlabel('Specialization')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
percentages = df["degree_field"].value_counts(normalize=True) * 100
percentages.plot(kind='bar')
plt.title('Degree field Percentages')
plt.xlabel('Degree')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
pct = pd.crosstab(df["specialisation"], df["gender"], normalize="index") * 100
ax = pct.plot(kind="bar", stacked=True, figsize=(8,5))

for container in ax.containers:
    labels = [f"{v:.1f}%" if v > 0 else "" for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="center")

plt.title("Gender Percentage by Specialization")
plt.xlabel("Specialization")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="status")
plt.title("Placement Status Count")
plt.xlabel("Placement Status")
plt.ylabel("Count")
plt.show()
```

```python
pct_gender = pd.crosstab(df["gender"], df["status"], normalize="index") * 100
ax = pct.plot(kind="bar", stacked=True, figsize=(8,5))

for container in ax.containers:
    labels = [f"{v:.1f}%" if v > 0 else "" for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="center")
plt.title("Placement Percentage by Gender and Specialization")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.legend(title="Status")
plt.xticks(rotation=0)
plt.show()
```

```python
pct_workex = pd.crosstab(df["workex"], df["status"], normalize="index") * 100
ax = pct.plot(kind="bar", stacked=True, figsize=(8,5))
for container in ax.containers:
    labels = [f"{v:.1f}%" if v > 0 else "" for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="center")
plt.title("Placement Percentage by Work Experience")
plt.xlabel("Work Experience")
plt.ylabel("Percentage")
plt.legend(title="Status")
plt.xticks(rotation=0)
plt.show()
```

```python
placed_df = df[df["status"] == "Placed"]

plt.figure(figsize=(8,5))
sns.histplot(placed_df["salary"], bins=20, kde=True)
plt.title("Salary Distribution of Placed Students")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x=placed_df["salary"])
plt.title("Boxplot of Salary")
plt.xlabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=placed_df, x="gender", y="salary")
plt.title("Salary by Gender")
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=placed_df, x="specialisation", y="salary")
plt.title("Salary by Specialisation")
plt.xlabel("Specialisation")
plt.ylabel("Salary")
plt.show()
```

```python
score_cols = ["seconday_educ_per", "high_second_educ_per", "degree_p", "etest_p", "mba_p"]

df[score_cols].hist(figsize=(12,8), bins=15, edgecolor="black")
plt.suptitle("Distribution of Academic Scores", y=1.02)
plt.show()
```

```python
numeric_cols = ["seconday_educ_per", "high_second_educ_per", "degree_p", "etest_p", "mba_p", "salary"]

plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
